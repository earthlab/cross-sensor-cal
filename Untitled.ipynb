```python
import matplotlib.pyplot as plt
import numpy as np

# Time values for the original advancement curve
t = np.linspace(0, 5.2, 50000)
advancement = np.exp(t) - 1

# Define the fork point where innovation splits off
fork_time = 4.5
fork_value = np.exp(fork_time) - 1

# Time values for the innovation curve (starting at the fork)
innovation_t = np.linspace(fork_time, 15, 5000)
innovation = fork_value + np.exp(0.5 * (innovation_t - fork_time)) - 1  # Slower exponential growth

# Plotting
plt.figure(figsize=(10, 10))

# Plot original advancement path
plt.plot(t, advancement, label='Advancement (original path)', color='orange', linewidth=2)

# Plot innovation path starting at the fork point
plt.plot(innovation_t, innovation, label='Innovation (disruptive path)', color='orangered', linewidth=2)

# Mark fork point
plt.plot(fork_time, fork_value, 'kx', markersize=10)
plt.axvline(fork_time, color='gray', linestyle='--', linewidth=1)
plt.text(fork_time + 0.1, fork_value - 5, 'Innovation Fork', fontsize=10)

# Axis labels and title
plt.title('Illustration of Advancement vs. Innovation', fontsize=14)
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import requests

flight_lines = [
    "NEON_D10_R10C_DP1_20210915_191104_radiance.h5",
    "NEON_D10_R10C_DP1_20210915_191530_radiance.h5",
    "NEON_D10_R10C_DP1_20210915_192104_radiance.h5",
    "NEON_D10_R10C_DP1_20210915_192523_radiance.h5",
    "NEON_D10_R10C_DP1_20210915_193014_radiance.h5",
]

base_url = "https://data.neonscience.org/data-files/DP1.30006.001/"

for file_name in flight_lines:
    url = base_url + file_name
    r = requests.get(url)
    if r.status_code == 200:
        with open(file_name, 'wb') as f:
            f.write(r.content)
        print(f"Downloaded: {file_name}")
    else:
        print(f"Failed: {file_name}  status {r.status_code}")
```
