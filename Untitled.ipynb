```python
from src.file_types import NEONReflectanceFile
from src.neon_to_envi import neon_to_envi, export_anc
from pathlib import Path

def run_flight_lines_to_envi(input_dir: str, output_dir: str, anc: bool = True):
    """
    Run neon_to_envi for all NEON reflectance H5 files in a directory,
    with progress reporting for Jupyter notebooks.
    """
    input_path = Path(input_dir)
    reflectance_files = NEONReflectanceFile.find_in_directory(input_path)
    
    if not reflectance_files:
        print(f"âŒ No NEON reflectance H5 files found in {input_dir}.")
        return

    print(f"âœ… Found {len(reflectance_files)} file(s) to process in {input_dir}\n")

    for i, neon_file in enumerate(reflectance_files, start=1):
        print(f"ðŸ“¦ [{i}/{len(reflectance_files)}] Converting: {neon_file.name}")
        try:
            neon_to_envi(
                images=[neon_file.file_path],
                output_dir=output_dir,
                anc=anc
            )
            print(f"âœ… Finished: {neon_file.name}\n")
        except Exception as e:
            print(f"âŒ Error processing {neon_file.name}: {e}\n")

    print("ðŸŽ‰ All files processed.")

# Example usage in notebook
run_flight_lines_to_envi("NIWO_2023_07", "NIWO_2023_07", anc=True)
```

```python
from pathlib import Path
from src.envi_download import download_neon_flight_lines
from src.neon_to_envi import neon_to_envi
from src.topo_and_brdf_correction import generate_config_json, topo_and_brdf_correction
from src.file_types import NEONReflectanceConfigFile

# Inputs
base_folder = Path("NIWO_2023_08")
site_code = "NIWO"
product_code = "DP1.30006.001"
year_month = "2023-08"
flight_lines = [
    "NEON_D13_NIWO_DP1_L019-1_20230815_directional_reflectance",
    "NEON_D13_NIWO_DP1_L020-1_20230815_directional_reflectance"
]

# Step 1: Download NEON flight lines
print("ðŸ“¥ Downloading NEON flight lines...")
download_neon_flight_lines(
    out_dir=base_folder,
    site_code=site_code,
    product_code=product_code,
    year_month=year_month,
    flight_lines=flight_lines
)
print("âœ… Download complete.\n")

# Step 2: Convert H5 to ENVI format (direct call to neon_to_envi)
print("ðŸ“¦ Converting H5 files to ENVI format...")
h5_files = list(base_folder.rglob("*.h5"))
if not h5_files:
    print("âŒ No .h5 files found for conversion.")
else:
    for i, h5_file in enumerate(h5_files, start=1):
        print(f"ðŸ”„ [{i}/{len(h5_files)}] Converting: {h5_file.name}")
        neon_to_envi(
            images=[str(h5_file)],
            output_dir=str(base_folder),
            anc=True
        )
        print(f"âœ… Finished: {h5_file.name}\n")
print("âœ… ENVI conversion complete.\n")

# Step 3: Generate configuration JSON
print("ðŸ“ Generating configuration JSON...")
generate_config_json(base_folder)
print("âœ… Config JSON generation complete.\n")

# Step 4: Apply topographic and BRDF corrections
print("â›°ï¸ Applying topographic and BRDF corrections...")
config_files = list(NEONReflectanceConfigFile.find_in_directory(base_folder, suffix="_config_envi.json"))
if config_files:
    for config_file in config_files:
        print(f"âš™ï¸ Applying corrections to: {config_file.file_path}")
        topo_and_brdf_correction(config_file.file_path)
    print("âœ… Corrections applied.\n")
else:
    print("âŒ No config files found. Skipping corrections.")

print("ðŸŽ‰ Pipeline complete!")
```
