# reading files from underflights

```python
import os
import pandas as pd
import geopandas as gpd
from pathlib import Path

# Define the root directory
root_dir = "/data-store/iplant/home/shared/earthlab/macrosystems/field-data/output/summer-2023-10cm-10k"

# Define GeoJSON path
geojson_path = "/data-store/iplant/home/shared/earthlab/macrosystems/field-data/analysis_ready_polygons/aop_polygons_1_24_2025_analysis_ready.geojson"

target_suffix = "pixel.csv"

df_list = []

# Loop through CSV files
for file_path in Path(root_dir).rglob("*.csv"):
    # Check if any parent directory contains "AOP"
    if any("AOP" in str(parent) for parent in file_path.parents):
        if file_path.name.endswith(target_suffix):
            try:
                df = pd.read_csv(file_path)
                df_list.append(df)
                print(f"Loaded: {file_path}")
            except Exception as e:
                print(f"Error reading {file_path}: {e}")
    else:
        continue  # Skip files not inside AOP directories

# Combine all CSVs
if df_list:
    combined_df_AOP = pd.concat(df_list, ignore_index=True)
    print(f"Combined DataFrame Shape: {combined_df_AOP.shape}")
else:
    print("No matching CSV files found inside AOP directories.")
    combined_df_AOP = None

# Load GeoJSON
geo_df = gpd.read_file(geojson_path)

# # Merge if IDs exist
# if combined_df is not None and "ID" in combined_df.columns and "ID" in geo_df.columns:
#     merged_df = geo_df.merge(combined_df, on="ID", how="inner")
#     print(f"Merged DataFrame Shape: {merged_df.shape}")

#     import ace_tools as tools
#     tools.display_dataframe_to_user(name="Merged Geo Data", dataframe=merged_df)
# else:
#     print("ID column missing or no CSVs loaded.")
```
