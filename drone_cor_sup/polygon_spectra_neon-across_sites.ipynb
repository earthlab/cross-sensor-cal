# reading files in 2024

```python
import os
import pandas as pd
import geopandas as gpd
from pathlib import Path

# Define the directory containing the CSV files
root_dir = "/data-store/iplant/home/shared/earthlab/macrosystems/processed_flight_lines"  # Your directory

# # Define the GeoJSON file path
# geojson_path = "/home/jovyan/data-store/cross-sensor-cal/aop_macrosystems_data_1_7_25.geojson"  # Your file

# Define target file pattern
target_suffix = "envi_resample_MicaSense_masked_spectral_data.csv"

# List to store DataFrames
df_list = []

# Recursively search and read CSV files in subdirectories
for file_path in Path(root_dir).rglob("*.csv"):
    if file_path.name.endswith(target_suffix):  # Check if filename ends with 'pixel.csv'
        try:
            df = pd.read_csv(file_path)
            df_list.append(df)
            print(f"Loaded: {file_path}")
        except Exception as e:
            print(f"Error reading {file_path}: {e}")

# Combine all CSV files into one DataFrame
if df_list:
    combined_df_2024 = pd.concat(df_list, ignore_index=True)
    print(f"Combined DataFrame Shape: {combined_df_2024.shape}")
else:
    print("No matching CSV files found.")
    combined_df_2024 = None

# # Load the GeoJSON file
# geo_df = gpd.read_file(geojson_path)

# # Merge if 'ID' exists in both
# if combined_df is not None and "ID" in combined_df.columns and "ID" in geo_df.columns:
#     merged_df = geo_df.merge(combined_df, on="ID", how="inner")
#     print(f"Merged DataFrame Shape: {merged_df.shape}")

#     # Display merged DataFrame
#     import ace_tools as tools
#     tools.display_dataframe_to_user(name="Merged Geo Data", dataframe=merged_df)
# else:
#     print("ID column missing in either CSVs or GeoJSON, or no CSVs loaded.")
```

# reading files from underflights

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract wavelengths from column names
# wavelengths = [float(col.split("_")[0]) for col in mean_spectra_other.columns]
wavelengths = [444.0, 475.0, 531.0, 560.0, 650.0, 668.0, 705.0, 717.0, 740.0, 862.0]
# Assuming your DataFrame is named mean_spectra_other
column_mapping = {f"MicaSense_band_{i+1}": wavelengths[i] for i in range(10)}
mean_spectra_other_all.rename(columns=column_mapping, inplace=True)
std_error_other_all.rename(columns=column_mapping, inplace=True)


```

```python
from matplotlib import cm
from matplotlib.colors import to_hex
import matplotlib.colors as mcolors
# Example categories from your data (tuples of class & location)
categories = mean_spectra_other_all.index

# Extract unique main classes
main_classes = sorted(set([cat[0] for cat in categories]))
print(main_classes)



# Use tab10 or any qualitative colormap for distinct base colors
base_cmap = plt.get_cmap('tab10')
base_colors = {cls: base_cmap(i) for i, cls in enumerate(main_classes)}



# For each main class, get subcategories (locations)
color_dict = {}
for cls in main_classes:
    subcats = [cat for cat in categories if cat[0] == cls]
    n_subcats = len(subcats)
    
    # Create lighter variations using colormap (e.g., lighter to darker)
    shades = np.linspace(0.4, 1.0, n_subcats)  # Adjust brightness range as needed
    
    base_color = np.array(mcolors.to_rgb(base_colors[cls]))
    
    for i, cat in enumerate(subcats):
        # Blend toward white (lighter)
        new_color = base_color * shades[i] + (1 - shades[i])
        color_dict[cat] = to_hex(new_color)

plt.figure(figsize=(12, 6))
for i, category in enumerate(categories):
    mean_values = mean_spectra_other_all.loc[category]
    std_error_values = std_error_other_all.loc[category]
    
    # Compute 95% confidence interval (CI = mean ± 1.96 * standard error)
    ci_upper = mean_values + 1.96 * std_error_values
    ci_lower = mean_values - 1.96 * std_error_values
    
    # Plot mean spectral curve
    plt.plot(wavelengths, mean_values, label=category, color=colors[i], marker='o')
    
    # Fill 95% confidence interval
    plt.fill_between(wavelengths, ci_lower, ci_upper, color=colors[i], alpha=0.2)

# Labels and legend
plt.xlabel("Wavelength (nm)")
plt.ylabel("Reflectance")
plt.title("Mean Spectral Signatures NEON AOP all sites full diamters by Cover Category with 95% CI")
plt.legend(title="Cover Category")
plt.grid()

# Show plot
plt.show()
plt.savefig("/home/jovyan/data-store/cross-sensor-cal/drone_cor_sup/NEON_AOP_all_sites_resampled_micasense_full_diam.png", dpi=300, bbox_inches='tight')
```
